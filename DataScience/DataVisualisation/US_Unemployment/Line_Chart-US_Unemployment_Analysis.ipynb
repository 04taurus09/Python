# Visual analysis on unemployment rate in US. The data has been made available by [The United States Bureau of Labor Statistics (BLS)](https://www.bls.gov), who surveys and calculates the monthly unemployment rate. The monthly unemployment rate is from January 1948 to August 2016, saved it as unrate.csv

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
unrate = pd.read_csv('unrate.csv')
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
unrate.head(12)
```

###### Data in the dataset is always from the first of every month with corresponding unemployment rate value in percentage.

```python
plt.plot(unrate['DATE'].head(12), unrate['VALUE'].head(12))
plt.xticks(rotation=90)
plt.xlabel('Months')
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment Trend, 1948')
plt.show()
```

###### Line chart depicts that unemployment trend of year 1948 has always varied, in more general there is pattern that every three months it has either gone up or gone down.

```python
fig = plt.figure(figsize=(12,7)) # figsize parameter is to specify size of figure

ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

x_values_1 = unrate['DATE'].head(12)
y_values_1 = unrate['VALUE'].head(12)

x_values_2 = unrate['DATE'][12:24]
y_values_2 = unrate['VALUE'][12:24]

ax1.plot(x_values_1, y_values_1)
ax1.set_title('Unemployment Trend, 1948') # adds title to ax1

ax2.plot(x_values_2, y_values_2)
ax2.set_title('Unemployment Trend, 1949') # adds title to ax2

plt.show()
```

```python
fig = plt.figure(figsize=(12,12))

for i in range(0,5):
    subplot = fig.add_subplot(5,1,i+1)
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    subplot.plot(subset['DATE'], subset['VALUE'])
    
plt.show()
```

```python
#extrating month from date and creating new column 'MONTH'
unrate['MONTH'] = unrate['DATE'].dt.month
unrate.head(24)

# Insted of creating multiple plots of each year, we can create only one plot with data of different year
fig = plt.figure(figsize=(12,6))
plt.plot(unrate[0:12]['MONTH'], unrate[0:12]['VALUE'],label='1948')
plt.plot(unrate[12:24]['MONTH'], unrate[12:24]['VALUE'],label='1949')
plt.legend(loc='upper left')
plt.title('Unemployment Trend,1948-49')
plt.xlabel('Month')
plt.ylabel('Rate')
plt.show()
```

```python
# Creating single plot containing data of five years
fig = plt.figure(figsize=(12,6))
year = ['1948','1949','1950','1951','1952']

for i in range(0,5):
    start_index = i*12
    end_index = (i*12)+12
    plt.plot(unrate['MONTH'][start_index:end_index], unrate['VALUE'][start_index:end_index], label=year[i])
plt.legend(loc='upper left')
plt.title('Unemployment Trend,1948-49')
plt.xlabel('Month')
plt.ylabel('Rate')
plt.show()
```

#### Few key points from the chart
* Unemployment rate was highest in October of 1949 and lowest in December of 1952
* Year 1950 has seen a gradual decrease in unemployment rate, with around 6.5% at start of year coming down to 4.2% during december
* In year 1949, unemployment rate started from aroung 4.2% with reaching to peak at around 8% in October and ending with 6.5% in December. Overall 1949, had highest unemployment rate.
